## Change Vaccinated

```python
def Vaccinated(Vaccinated):
  if Vaccinated == 1:
    return 'Vaccinated'
  else:
    return 'Not Vaccinated'
```

```python
pet['Vaccinated'] = pet['Vaccinated'].apply(Vaccinated)
```

## Change HealthCondition

```python
def HealthCondition(HealthCondition):
  if HealthCondition == 1:
    return 'Healthy'
  else:
    return 'Not Healthy'
```

```python
pet['HealthCondition'] = pet['HealthCondition'].apply(HealthCondition)
```

## Change PreviousOwner

```python
def PreviousOwner(PreviousOwner):
  if PreviousOwner == 1:
    return 'Yes'
  else:
    return 'No'
```

```python
pet['PreviousOwner'] = pet['PreviousOwner'].apply(PreviousOwner)
```

## Change AdoptionLikelihood

```python
def AdoptionLikelihood(AdoptionLikelihood):
  if AdoptionLikelihood == 1:
    return 'NotAdoption'
  else:
    return 'Adoption'
```

```python
def AdoptionLikelihood(AdoptionLikelihood):
  if AdoptionLikelihood == 1:
    return '1'
  else:
    return '1'
```

```python
pet['AdoptionLikelihoodHuman'] = pet['AdoptionLikelihood'].apply(AdoptionLikelihood)
```

```python
pet['AdoptionLikelihood'] = pet['AdoptionLikelihood'].apply(AdoptionLikelihood)
```

# Analysis

# Soal 1: Menghitung Statistik Dasar

### 1.1 Hitung rata-rata, median, dan standar deviasi untuk kolom AgeMonths.



```python
# Soal 1.1: Menghitung statistik dasar untuk AgeMonths
mean_age = pet['AgeMonths'].mean()
median_age = pet['AgeMonths'].median()
std_age = pet['AgeMonths'].std()

print(f"Rata-rata AgeMonths: {mean_age}")
print(f"Median AgeMonths: {median_age}")
print(f"Standar Deviasi AgeMonths: {std_age}")
```

```python
pet['AgeMonths'].describe()
```

#### 1.1.2 Buat histogram untuk kolom AgeMonths.

```python
plt.figure(figsize=(10, 6))
plt.hist(pet['AgeMonths'], bins=20, color='lightcoral', edgecolor='black')
plt.xlabel('Age (Months)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Histogram of Pet Ages (Months)', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

### 1.2 Hitung rata-rata, median, dan standar deviasi untuk kolom WeightKg.

```python
# Soal 1.2: Menghitung statistik dasar untuk WeightKg
mean_weight = pet['WeightKg'].mean()
median_weight = pet['WeightKg'].median()
std_weight = pet['WeightKg'].std()

print(f"Rata-rata WeightKg: {mean_weight}")
print(f"Median WeightKg: {median_weight}")
print(f"Standar Deviasi WeightKg: {std_weight}")
```

```python
pet['WeightKg'].describe()
```

#### 1.2.1 Buat histogram untuk kolom WeightKg.

```python
plt.figure(figsize=(10, 6))
plt.hist(pet['WeightKg'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Weight (Kg)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Histogram of Pet Weights', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

# Soal 4: Analisis Hubungan Antar Kolom


### 4.1 Buat scatter plot untuk AgeMonths dan WeightKg.



```python
# Define colors for each pet type
colors = {'Dog': 'blue', 'Cat': 'green', 'Rabbit': 'red', 'Bird': 'orange'}

plt.figure(figsize=(10, 6))

# Plot scatter plot for each pet type
for pet_type in colors.keys():
    subset = pet[pet['PetType'] == pet_type] # Changed 'Type' to 'PetType'
    plt.scatter(subset['AgeMonths'], subset['WeightKg'],
                c=colors[pet_type], label=pet_type, alpha=0.6, edgecolors='w', s=100)

plt.title('Scatter plot antara AgeMonths dan WeightKg', fontsize=15)
plt.xlabel('Age (Months)', fontsize=12)
plt.ylabel('Weight (Kg)', fontsize=12)
plt.legend(title='Pet Type')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Soal 4.2: Menghitung korelasi antara AgeMonths dan WeightKg

```python
correlation = pet[['AgeMonths', 'WeightKg']].corr().iloc[0, 1]
print(f"Korelasi antara AgeMonths dan WeightKg: {correlation}")
```
